## Padronização das datas

```python
%sql
-- Limpar e estruturar a tabela de Fatos (Vendas) na Camada Prata.
-- Criar nova coluna para a data padronizada
-- ==============================================================================

CREATE OR REPLACE TABLE lh_nautical.lh_nautical_db.slv_vendas AS
WITH cte_vendas_deduplicadas AS (
    SELECT DISTINCT * FROM lh_nautical.lh_nautical_db.vendas_2023_2024
    WHERE id IS NOT NULL 
      AND id_client IS NOT NULL 
      AND id_product IS NOT NULL
)
SELECT 
    id AS id_sale,
    id_client,
    id_product,
    qtd,
    total AS total_num,
    COALESCE(
        TRY_TO_DATE(sale_date, 'yyyy-MM-dd'), 
        TRY_TO_DATE(sale_date, 'dd-MM-yyyy')  
    ) AS sale_date_clean
    
FROM cte_vendas_deduplicadas;

-- amostra pra analise
SELECT 
    sale_date_clean 
FROM lh_nautical.lh_nautical_db.slv_vendas
LIMIT 10;
```
